```python
import numpy as np
import pandas as pd
import sys
sys.path.append('D:\Desktop\Вся прога на питоне\Задача по курсовой')
from functions import ONP_Adagrad,ONP,to_exel,train_test_val,NN_weight,to_exel_GB
```

```python
train = pd.read_csv('D:/Desktop/Вся прога на питоне/Задача по курсовой/mtsgrvmgn_trn.csv')
test = pd.read_csv('D:/Desktop/Вся прога на питоне/Задача по курсовой/mtsgrvmgn_tst.csv')
valid = pd.read_csv('D:/Desktop/Вся прога на питоне/Задача по курсовой/mtsgrvmgn_vld.csv')
```

```python
columns = ['H1_8']
```

```python
X_train,X_test,X_valid,y_train,y_test,y_valid = train_test_val(train=train,test=test,valid = valid,columns=columns)
top_750_importance = NN_weight(X_train=X_train,X_valid=X_valid,y_train=y_train,y_valid=y_valid,
                               batch_size=64,input_dim=X_train.shape[1],output_dim=len(columns),
                               learning_rate=0.001,num_epochs=50,patience=10,graph=True)
X_train_top_750 = top_750_importance
to_exel_GB(file_name='All_GB_H1.xlsx',
        n_coloumns=4,n_iter=1,X_train=X_train,y_train=y_train,
        X_valid=X_valid,y_valid=y_valid,X_test=X_test,y_test=y_test,n_estimators=256,max_depth=5,subsample=1)
```
