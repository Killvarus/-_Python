```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m_values = [1, 2, 3, 4, 5] 
n_values = [1, 2, 3, 4, 5]  
phi = np.linspace(0, 2 * np.pi, 1000)  
```

```python
fig, axes = plt.subplots(max(m_values), max(n_values), figsize=(5*max(m_values) // 2, 4 * max(n_values) // 2))

for i, m in enumerate(m_values):
    for j, n in enumerate(n_values):
        phi = np.linspace(0, 2 * np.pi * (max(m_values)), 1000)
        k = (i+1)/(j+1)
        r = np.cos(k * phi)
        x = r * np.cos(phi)
        y = r * np.sin(phi)
        ax = axes[i][ j]
        ax.plot(x, y, label=f"k={i+1}/{j+1}")
        ax.set_aspect('equal')
        ax.axis('off')

for j, n in enumerate(n_values):
    axes[0, j].set_title(f'n = {n}', fontsize=12) 

for i, m in enumerate(m_values):
    axes[i, 0].set_title( f'm = {m}', fontsize=12,rotation = 90,x=-0.1,y=0.3) 
 
    
plt.show()
```
