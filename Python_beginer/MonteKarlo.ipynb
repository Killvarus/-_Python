```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
КОЛ_ВО = 100000

np.random.seed(42)
x_points = np.random.uniform(-1, 1, КОЛ_ВО)
y_points = np.random.uniform(-1, 1, КОЛ_ВО)

попало = x_points**2 + y_points**2 <= 1
колво_внутри = np.sum(попало)
площадь = 4 * колво_внутри / КОЛ_ВО
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x_points[попало], y_points[попало], color='red', s=1, label='Внутри')
plt.scatter(x_points[~попало], y_points[~попало], color='black', s=1, label='Снаружи')
круг = plt.Circle((0, 0), 1, color='green', fill=False, linewidth=2)
plt.gca().add_artist(круг)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.legend()
plt.title(f"Метод Монте-Карло: оценка площади = {площадь:.4f}, точное значение = {np.pi:.4f}")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
