```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
N = 10000  
a, b = -4, 4  
bin_width = 0.2  
bins = np.arange(a, b, bin_width)  
data = np.random.normal(loc=0, scale=1, size=N)
x = np.linspace(a, b, 1000)
pdf = norm.pdf(x, loc=0, scale=1)  
pdf_scaled = pdf * N * bin_width  
```

```python
plt.hist(data, bins=bins, density=False, alpha=0.6, color='skyblue', edgecolor='black', label='Гистограмма')
plt.plot(x, pdf_scaled, 'r-', linewidth=2)
plt.xlabel('Значение случайной величины')
plt.ylabel('Частота')
plt.title('Плотность вероятности')
plt.legend()
plt.show()
```
