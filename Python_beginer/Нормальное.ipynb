```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
mu = 0 
сигмы = [0.5, 1, 2] 
цвета = ['blue', 'green', 'red']
x = np.linspace(-10, 10, 1000)
plt.figure(figsize=(8, 6))
```

```python
for sigma in сигмы:
    y = norm.pdf(x, mu, sigma)
    plt.plot(x, y, label=f'σ = {sigma}')
plt.xlabel('x')
plt.ylabel('Плотность вероятности')
plt.legend(title='Сигма')
plt.title('Графики нормального распределения для разных значений сигмы')
plt.grid(True)
plt.show()
```
